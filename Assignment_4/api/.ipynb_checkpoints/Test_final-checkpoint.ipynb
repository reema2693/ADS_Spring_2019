```python
from flask import Flask, render_template, redirect, url_for, request
import requests as rq
import json
from watson_developer_cloud import ToneAnalyzerV3
import re

def getResponse(cntry, noOfSongs ):
    # To store the list of tracks and lyrics
    tracks = []
    lyrics = []
    url = []
    
    #base url
    base_url = 'https://api.musixmatch.com/ws/1.1/'

    #apikey
    api_key = '&apikey=5d05ee18c704d1bfa04ec7643f980e04'

    #format url to form the json response
    format_url = '?format=json&callback=callback'

    #country code fetched from web page
    country = cntry

    #input paramter to fetch the tracks on musixmatch
    parameter = 'chart.tracks.get'
    chart_name= 'top'
    page = 1

    #Total  number of songs
    page_size = noOfSongs
    f_has_lyrics = 1

    #request url 
    api_call = base_url+parameter+format_url+'&chart_name='+chart_name+'&page='+str(page)+'&page_size='+str(page_size)+'&country='+country+'&f_has_lyrics='+str(f_has_lyrics)+api_key
#    print(api_call)

    #response
    response = rq.get(api_call)
    data = response.json()
    data = data['message']['body']
      
    for i in data['track_list']:
        for j in i.values():
            tracks.append(j['track_name'])
#    print(tracks)

    #input paramter to fetch the lyrics on musixmatch
    ly_param = 'track.lyrics.get'

    for k in data['track_list']:
        for l in k.values():
            api_lyrics = base_url+ly_param+format_url+'&track_id='+str(l['track_id'])+api_key
            ly_response = rq.get(api_lyrics)
            url.append(api_lyrics)
            new_data = ly_response.json()
            new_data = new_data['message']['body']
            for m in new_data.values():
                lyrics.append(m['lyrics_body'])     
    #             lyrics = m['lyrics_body']
    #    print(lyrics)
    #    print(" ")
    #         print(new_data.values())
    
    final_result = dict(zip(tracks, url))
    return final_result

################################################
def getLyrics(url):
    lyrics1=[]
    ly_response = rq.get(url)
    new_data = ly_response.json()
    new_data = new_data['message']['body']
    for m in new_data.values():
        lyrics1.append(m['lyrics_body'])     
    return lyrics1

################################################

def toneAnalyzer(lyrics):
    tone_analyzer = ToneAnalyzerV3(
        version='2017-09-21',
        iam_apikey='MnPSf72JUQ8-8Ik1q5yTCq0ew5Lp_Qw2vV3Unu9SK-MN',
        url='https://gateway-wdc.watsonplatform.net/tone-analyzer/api'
    )
    fullStr = ' '.join(lyrics)
    line = fullStr.replace('\n',' ')
    words = line.split("...")

    tone_analysis = tone_analyzer.tone(
        {'text': words[0]},
        'application/json'
    ).get_result()
    tone= json.dumps(tone_analysis, indent=2)
    return tone

################################################
def tone_final(lyric_tone):
    def replaceMultiple(mainString, toBeReplaced, newString):
        # Iterate over the strings to be replaced
        for elem in toBeReplaced :
            # Check if string is in the main string
            if elem in mainString :                        
                # Replace the string
                mainString = mainString.replace(elem, newString)
        return  mainString  
    def find_tone(lyric_tone):
        tone_string = replaceMultiple(lyric_tone, ['\n', '{', '}','[',']',' ','document_tone','tones','""','"','::'] , "")
        tone_list = tone_string.split(',')
        print(tone_list)
        if '' in tone_list :
            tone_list.remove('')
        tone_list_final=[a.split(':',1) for a in tone_list]
        print(tone_list_final)
        tone_dict={}
        for key, val in tone_list_final:
            tone_dict.setdefault(key, []).append(val)
        maxx = max(tone_dict.get('score'))
        for k, v in tone_dict.items():
            aaa = len(v)
            for i in range(aaa):
                if v[i] == maxx:
                    tone_i = i
        b = tone_dict.get('tone_name')
        tone_final=b[tone_i]
        return tone_final
    
    num = lyric_tone.find('"score"')
    if num > 0:
        num1 = lyric_tone.find('"sentences_tone"')
        if num1 <0 :
            tone_final = find_tone(lyric_tone)
            
        elif num>num1:
            numb=lyric_tone.find("sentence_id")
            lyric_tone =lyric_tone[numb:]
            tone_string = replaceMultiple(lyric_tone, ['\n', '{', '}','[',']',' ','document_tone','tones','""','"','::'] , " ")
            tone_string = re.sub(' +',' ',tone_string)
            tone_string = tone_string.replace(', :',',')
            tone_list = tone_string.split(', ')
            tone_list_final=[a.split(':',1) for a in tone_list]
            len_list= len(tone_list_final)
            del_list = []
            for i in range(len_list):
                b = len(tone_list_final[i])
                if b == 1:
                    del_list.append(i)
            for c in del_list:
                del tone_list_final[c]
            tone_dict={}
            for key, val in tone_list_final:
                tone_dict.setdefault(key, []).append(val)
            tone_sent=[]
            for x in tone_dict['tone_name ']:
                if x == 'Anger' or x =='Fear' or x == 'Sadness' or x == 'Tentative':
                    x ='Sad'
                    tone_sent.append(x)
                else:
                    x ='Happy'
                    tone_sent.append(x)
            a = tone_sent.count('Happy')
            b = tone_sent.count('Sad')
            if a>b:
                tone_final ='Happy'
            elif a<b:
                tone_final ='Sad'
            else:
                tone_final ='Happy'
        else:
            if num1>0:
                lyric_tone = lyric_tone[:num1]
                tone_final = find_tone(lyric_tone)
    else:
        tone_final=("Not able to find the Tone")
    return tone_final

################################################      
   
app = Flask(__name__)

@app.route('/')
def index():
    country = ['UK','Canada','Australia']
    no = ['5','10','15','20']
    return render_template('index.html', countries = country, noOfSongs=no)

@app.route("/result" , methods=['GET','POST'])
def result():
    if request.method =='POST':
        result = request.form
        coun = request.form['country']
        num = request.form['num']
        tracks = getResponse(coun,num)
        return render_template("songslist.html",tracks = tracks)
    
@app.route("/single" , methods=['GET','POST'])
def single():
        if request.method =='POST':
            single=request.form
            url=request.form['SongName']
            url_value = getLyrics(url)
            lyric_tone = toneAnalyzer(url_value)
            tone = tone_final(lyric_tone)
            if tone == 'Anger' or tone =='Fear' or tone == 'Sadness' or tone == 'Tentative':
                tone_assign = 'Sad'
            elif tone == 'Not able to find the Tone':
                tone_assign = 'Not able to find the Tone'
            else:
                tone_assign = 'Happy'
            return render_template("single.html", url_value = url_value,tone=tone,tone_assign=tone_assign)
app.run(debug = False)
```
