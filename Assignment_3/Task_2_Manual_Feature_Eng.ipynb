# Task 2: Manual Feature Engineering

```python
import nltk
import csv
import os
import pandas as pd
import numpy as np
from xlwt import Workbook
from statistics import mean

grade_name=[]
risk_factor=[]
inst_rate=[]
intrest_range=[]

mean_data=[]
actual_mean1=[]
actual_sum1=[]
actual_std1=[]
actual_max1=[]
actual_min1=[]

#reading data from csv file
df=pd.read_csv(os.getcwd()+'/cleaned_data.csv')
df=df.fillna(" ")

#Saving data of grade column
for i in df.columns:
    i
    for date_file in df['grade']:
        grade_name.append(date_file)

#Saving data of Intrest rate column
for i in df.columns:
    i
    for date_file in df['int_rate']:
        inst_rate.append(date_file)
        
#assigning risk factor on the basis of grade        
for grade_list in grade_name:
    if(grade_list == 'A'):
        risk_factor.append('Very Low')
    elif(grade_list == 'B' or grade_list == 'C'):
        risk_factor.append('Low')
    elif(grade_list == 'D'):
        risk_factor.append('Moderate')
    elif(grade_list == 'E' or grade_list == 'F'):
        risk_factor.append('High')
    elif(grade_list == 'G'):
        risk_factor.append('Very High')
        
#assigning range on the basis of intrest rate
for inst_list in inst_rate: 
    if(inst_list > 5 and inst_list <= 10):
        intrest_range.append('5 To 10')
    elif(inst_list > 10 and inst_list <= 15):
        intrest_range.append('10 To 15')
    elif(inst_list > 15 and inst_list <= 20):
        intrest_range.append('15 To 20')
    elif(inst_list > 20 and inst_list <= 25):
        intrest_range.append('20 To 25')
    elif(inst_list > 25 and inst_list <= 30):
        intrest_range.append('25 To 30')

#Calculating the mean,sum,standard deviation,maximum and minimun of intrest rate based on the grades
        
grouped = df.groupby('grade')
mean_value = grouped['int_rate'].mean()
actual_mean=mean_value.tolist()

sum_value = grouped['int_rate'].sum()
actual_sum= sum_value.tolist()

std_value = grouped['int_rate'].std()
actual_std= std_value.tolist()

max_value = grouped['int_rate'].max()
actual_max=max_value.tolist()

min_value = grouped['int_rate'].min()
actual_min=min_value.tolist()

for grade_list in grade_name:
    if(grade_list == 'A'):
        actual_mean1.append(actual_mean[0])
        actual_sum1.append(actual_sum[0])
        actual_std1.append(actual_std[0])
        actual_max1.append(actual_max[0])
        actual_min1.append(actual_min[0])
    elif(grade_list == 'B'):
        actual_mean1.append(actual_mean[1])
        actual_sum1.append(actual_sum[1])
        actual_std1.append(actual_std[1])
        actual_max1.append(actual_max[1])
        actual_min1.append(actual_min[1])
    elif(grade_list == 'C'):
        actual_mean1.append(actual_mean[2])
        actual_sum1.append(actual_sum[2])
        actual_std1.append(actual_std[2])
        actual_max1.append(actual_max[2])
        actual_min1.append(actual_min[2])
    elif(grade_list == 'D'):
        actual_mean1.append(actual_mean[3])
        actual_sum1.append(actual_sum[3])
        actual_std1.append(actual_std[3])
        actual_max1.append(actual_max[3])
        actual_min1.append(actual_min[3])
    elif(grade_list == 'E'):
        actual_mean1.append(actual_mean[4])
        actual_sum1.append(actual_sum[4])
        actual_std1.append(actual_std[4])
        actual_max1.append(actual_max[4])
        actual_min1.append(actual_min[4])
    elif(grade_list == 'F'):
        actual_mean1.append(actual_mean[5])
        actual_sum1.append(actual_sum[5])
        actual_std1.append(actual_std[5])
        actual_max1.append(actual_max[5])
        actual_min1.append(actual_min[5])
    elif(grade_list == 'G'):
        actual_mean1.append(actual_mean[6])
        actual_sum1.append(actual_sum[6])
        actual_std1.append(actual_std[6])
        actual_max1.append(actual_max[6])
        actual_min1.append(actual_min[6])
                
#Inserting calculated values in the columns

risk_fac = pd.Series(risk_factor)
df.insert(29, 'Risk Factor', risk_fac)

inst_range = pd.Series(intrest_range)
df.insert(30, 'Intrest Rate Range', inst_range)

csv_mean = pd.Series(actual_mean1)
df.insert(31, 'Mean(Interest_Rate)', csv_mean)

csv_sum = pd.Series(actual_sum1)
df.insert(32, 'Sum(Interest_Rate)', csv_sum)

csv_std = pd.Series(actual_std1)
df.insert(33, 'Std(Interest_Rate)', csv_std)

csv_max = pd.Series(actual_max1)
df.insert(34, 'Max(Interest_Rate)', csv_max)

csv_min = pd.Series(actual_min1)
df.insert(35, 'Min(Interest_Rate)', csv_min)

df.to_csv(os.getcwd()+'/MF_Data.csv',index = False)
```
