```python
import glob
from PyPDF2 import PdfFileWriter, PdfFileReader

#This Function merges all the 4 PDF
def merger(output_path, input_paths):
    pdf_writer = PdfFileWriter()
 
    for path in input_paths:
        pdf_reader = PdfFileReader(path)
        for page in range(pdf_reader.getNumPages()):
            pdf_writer.addPage(pdf_reader.getPage(page))
 
    with open(output_path, 'wb') as fh:
        pdf_writer.write(fh)
        
#Change the path here according to your machine        
if __name__ == '__main__':
    paths = glob.glob('/Users/rajsharavan/Desktop/Python/Sam*.pdf')
    paths.sort()
    merger('/Users/rajsharavan/Desktop/Python/Pdf_merger.pdf', paths)
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import collections
from tika import parser
import csv
import sys
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.converter import XMLConverter, HTMLConverter, TextConverter
from pdfminer.layout import LAParams
import io

#This Function converts the Merged PDF to text
def pdfparser(data):

    fp = open(data, 'rb')
    rsrcmgr = PDFResourceManager()
    retstr = io.StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
# Create a PDF interpreter object.
    interpreter = PDFPageInterpreter(rsrcmgr, device)
# Process each page contained in the document.
    for page in PDFPage.get_pages(fp):
        interpreter.process_page(page)
        data =  retstr.getvalue()
    print(data)
    #print(type(data))
    stop_words = set(stopwords.words('english'))   
    word_tokens = word_tokenize(data)
    words = [w for w in word_tokens if w.isalpha()]
    filtered_sentence = []

#Checks the overall words and filters the stop_words    
    for w in words: 
        #w.lower()
        #print(w.lower())
        if w.lower() not in stop_words: 
            filtered_sentence.append(w.lower())
    print(len(filtered_sentence))
    #print(filtered_sentence)
        
    def get_count(word_count_tuple):
        return word_count_tuple[1]

    word_counts = collections.Counter(filtered_sentence)

# This function sorts the words based on the number of times repeated and saves them in a csv file
    def print_top():
        word_count = word_counts
        items = sorted(word_count.items(),key=get_count,reverse=True)
        with open('/Users/rajsharavan/Desktop/Python/WordList.csv','w') as out:
            csv_out=csv.writer(out)
            #csv_out.writerow('Name')
            for row in items:
                csv_out.writerow(row)

    print_top()
```

```python
#By calling this function, it does all the jobs(PDF to text, Filtering Stop_words and creating a CSV file)
path = ('/Users/rajsharavan/Desktop/Python/Pdf_merger.pdf')
pdfparser(path)
```
